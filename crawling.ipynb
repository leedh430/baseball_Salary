```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
chrome_options = webdriver.ChromeOptions() 
chrome_options.add_argument('--headless') #내부 창을 띄울 수 없으므로 설정 
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage') 
chromeDriver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
chromeDriver.get("http://www.statiz.co.kr/stat.php?opt=0&sopt=0&re=1&ys=2018&ye=2018&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR&o2=OutCount&de=1&lr=0&tr=&cv=&ml=1&sn=1000&si=&cn=")
html = chromeDriver.page_source
bspit_18 = BeautifulSoup(html, 'html.parser')
pit_18 = bspit_18.find_all("table")[1]

chromeDriver.get("http://www.statiz.co.kr/stat.php?mid=stat&re=1&ys=2019&ye=2019&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR&o2=OutCount&de=1&lr=0&tr=&cv=&ml=1&pa=0&si=&cn=&sn=1000")
html = chromeDriver.page_source
bspit_19 = BeautifulSoup(html, 'html.parser')
pit_19 = bspit_19.find_all("table")[1]

chromeDriver.get("http://www.statiz.co.kr/stat.php?opt=0&sopt=0&re=1&ys=2020&ye=2020&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR&o2=OutCount&de=1&lr=0&tr=&cv=&ml=1&sn=1000&si=&cn=")
html = chromeDriver.page_source
bspit_20 = BeautifulSoup(html, 'html.parser')
pit_20 = bspit_20.find_all("table")[1]
```

```python
column_pit = ["순","이름","팀","WAR","출장","완투","완봉","선발","승","패","세","홀드","이닝","실점","자책","타자","안타","2타","3타",
              "홈런","볼넷","고4","사구","삼진","보크","폭투","ERA","FIP","WHIP","ERA+","FIP+","WAR","WPA"]

df_18 = pd.DataFrame(columns=column_pit)
templen = len(pit_18.find_all("tr"))
 
for i in range(2, templen):
  tempTr = pit_18.find_all("tr")[i]
  if(tempTr.find("th") is not None):
    continue
  row = {}
  column_idx = 0
  for j in range(len(column_pit)):
    tempTd = tempTr.find_all("td")[j].text
    if(tempTd == "  " and j > 0):
      continue
    row[column_pit[column_idx]] = tempTd
    column_idx += 1
  df_18 = df_18.append(row,ignore_index=True)


df_19 = pd.DataFrame(columns=column_pit)
templen = len(pit_19.find_all("tr"))
 
for i in range(2, templen):
  tempTr = pit_19.find_all("tr")[i]
  if(tempTr.find("th") is not None):
    continue
  row = {}
  column_idx = 0
  for j in range(len(column_pit)):
    tempTd = tempTr.find_all("td")[j].text
    if(tempTd == "  " and j > 0):
      continue
    row[column_pit[column_idx]] = tempTd
    column_idx += 1
  df_19 = df_19.append(row,ignore_index=True)

df_20 = pd.DataFrame(columns=column_pit)
templen = len(pit_20.find_all("tr"))
 
for i in range(2, templen):
  tempTr = pit_20.find_all("tr")[i]
  if(tempTr.find("th") is not None):
    continue
  row = {}
  column_idx = 0
  for j in range(len(column_pit)):
    tempTd = tempTr.find_all("td")[j].text
    if(tempTd == "  " and j > 0):
      continue
    row[column_pit[column_idx]] = tempTd
    column_idx += 1
  df_20 = df_20.append(row,ignore_index=True)
```

```python
column = ["이름","WAR","출장","완투","완봉","선발","승","패","세","홀드","이닝","실점","자책","타자","안타","2타","3타",
              "홈런","볼넷","고4","사구","삼진","보크","폭투","ERA","FIP","WHIP","ERA+","FIP+","WAR","WPA"]
```

```python
df_sub = pd.DataFrame(columns = column)
idx=0
for i in range(len(df_19)):
  candi_19 = df_19.loc[i]
  for j in range(len(df_18)):
    candi_18 = df_18.loc[j]
    if candi_19["이름"] == candi_18["이름"]:
      temp = [candi_19["이름"]]
      for k in range(3,len(candi_19)):
        temp.append(float(candi_19[k])- float(candi_18[k]))
      df_sub.loc[idx] = temp
      idx+=1
df_sub.to_csv('test.csv')

df_sub = pd.DataFrame(columns = column)
idx=0
for i in range(len(df_19)):
  candi_19 = df_19.loc[i]
  for j in range(len(df_20)):
    candi_20 = df_20.loc[j]
    if candi_19["이름"] == candi_20["이름"]:
      temp = [candi_20["이름"]]
      for k in range(3,len(candi_19)):
        temp.append(float(candi_20[k])- float(candi_19[k]))
      df_sub.loc[idx] = temp
      idx+=1

df_sub.to_csv('1920.csv')
```
